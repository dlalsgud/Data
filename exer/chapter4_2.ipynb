```python
# cv2, numpy, matplotlib.pyplot

import cv2 #컴퓨터비전용 라이브러리
import numpy as np #행렬 연산
import matplotlib.pyplot as plt #시각화 라이브러리
import math
```

```python
img = cv2.imread("images/Lenna.png", cv2.IMREAD_GRAYSCALE) #RGB
h,w = img.shape
print(h,w)
```

```python
#이중 반복문을 활용하여, 영상(배열)의 각 픽셀에 접근하고

ex2_result = np.zeros((h,w)) #초기화 되어 있다.
for i in range(0,h):
    for j in range(0,w):
        if img[i,j] * 1.7 <= 255:
            ex2_result[i,j] = math.trunc(img[i,j] * 1.7)
        else:
            ex2_result[i,j] = math.trunc(img[i,j] * 1.7) - 255 - 1 #256->0 257->1
```

```python
plt.imshow(ex2_result, cmap='gray')
```
